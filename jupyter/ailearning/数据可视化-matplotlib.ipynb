### numpy.random.rand() 生成numpy array,参数为数组维度

```python
import matplotlib.pyplot as plt
from numpy.random import rand

x = rand(200)
y = rand(200)
size = rand(200) * 30
color = rand(200)

plt.scatter(x,y,size,color)
plt.colorbar()
```

##  figure() 多子图

```python
t = np.linspace(0,2*pi,50)
x = np.sin(t)
y = np.cos(t)
plt.figure()
plt.plot(t,x)
plt.figure()
plt.plot(t,y)
```

## subplot() 多子图

```python

import matplotlib.pyplot as plt
from math import pi
import numpy as np
from matplotlib.patches import Polygon
t = np.linspace(0,2*pi,50)
x = np.sin(t)
y = np.cos(t)
ax1 = plt.subplot(221)
ax1.margins(0.5)     
ax1.plot(t,x)
ax1.set_title("aa")

ax2 = plt.subplot(231)
ax2.plot(t,y)
ax2.set_title("bb")
plt.show()
```
